# Evaluating Impact of Stirring of Espresso on Total Dissolved Solids (TDS)

There have been several studies of TDS measurement in the context of precision of espresso devices. These studies have adopted procedures for stirring the espresso upfront using a spoon, specifying a number of stirs. The question this study looks to evaluate is the impact stirring methodology has on measuring TDS in espresso.

## Measuring TDS

There are three methods with which TDS will be measured for this experiment, they are as follows:

1. Stir espresso with spoon, sample, cool, measure without filtering
2. Stir espresso with stir plate, sample, cool, measure without filtering
3. Weigh dissolved solids as described [here][4]

The third method will be performed for every shot, which is intended as more accurate measure of the true TDS. while the sampling order of using a spoon to stir vs a stir plate will be decided up front using a Random Number Generator (RNG) to avoid measurement errors caused by sampling sequentially (getting tired, getting sloppy, etc).

### Syringe Reuse

Syringes are made up a significant amount of plastic, and perform this experiment to the standards found in a lab would require a lot of syringes used once then thrown out. This wasteful for a measurement that has low impact (IE good vs great coffee compared to say telling if a molecule is a drug or not). With that in mind syringes will be reused in a way that attempts to produce defensible values while avoiding too much waste. 

The syringes that collect the samples from the stirred coffee will be reused for the unfiltered TDS testing as well as collecting the sample for measuring the 'true' dissolved solids via evaporation[[4][4]].

There is an additional syringe used to tare the refractometer with distilled, which will be reused through out.

1. Wait 5 minutes for cooling
1. For three separate 1.0ml samples, perform the following steps
    1. Clean refractometer sensor, tare using distilled water
    1. Draw up 1.0ml of liquid
    1. Dipose of first 4 drops
    1. Fill refractometer sensor area
    1. Collect TDS and Temperature (from refractometer) every 30 seconds until TDS converges three times in a row or 10 times, whichever comes first.
1. Place set aside sample with ~10g into oven at 255F (123.89C)
1. Once sample is dry, compute TDS, record. Done afterwards


###  Stirring With a Spoon

1. Stir shot with a small spoon ten times clockwise and ten times counterclockwise

### Stirring With Stir Plate

1. Add clean stir rod
1. Turn on stir plate to lowest speed that creates clean vortex without splatter
1. Wait 10 seconds
1. Turn off stir plate, wait for liquid to stop spinning.

```python
# Download CSV, won't work on windozes
with NamedTemporaryFile(suffix=".csv") as temp:
    with open(temp.name, "wb") as ofs:
        resp = requests.get(DATA_DOWNLOAD_URL, stream=True)
        for chunk in resp.iter_content():
            ofs.write(chunk)
    df = pd.read_csv(temp.name)
# Clean up the CSV file.
for char in [' ', '(', ')', "%", "/", "\\", '"']:
    df.columns = df.columns.str.strip().str.lower().str.replace(char, '')
df = df[df["grinder"].astype(bool)][df["grindersetting"].astype(bool)]
# Only look at data with TDS values and ratings
df = df.assign(extraction_yield = (df.tds * df.output) / df.coffeegrams)
df.date = pd.to_datetime(df.date, infer_datetime_format=True, utc=True)
df.roastdate = pd.to_datetime(df.roastdate, infer_datetime_format=True, utc=True)
df = df.assign(days_off_roast = df.date - df.roastdate)

evaporated_df = df[~df["containerweight+driedsolids"].isna()]
evaporated_df = evaporated_df.assign(
    tds=np.abs(
        (evaporated_df["containerweight+driedsolids"] - evaporated_df["containerweight"]) / evaporated_df.liquidweight * 100
    )
)
evaporated_df = evaporated_df.assign(extraction_yield = (evaporated_df.tds * evaporated_df.output) / evaporated_df.coffeegrams)
df.drop(df[df.tds.isna()].index, inplace=True)
```

## Hypothesis #2: Unfiltered TDS provides ranks ordering

While unfiltered measurements of espresso does not provide the most precise reading [[1](https://socraticcoffee.com/2015/07/measuring-total-dissolved-solids-a-refractometer-comparison-part-iii-espresso-filters/)], rank ordering is what matters for dialing in coffee. Despite the lack of precise, does measuring the unfiltered espresso produce a meaningful rank ordering? 

```python
for run in df.run.unique():
    taus = {}
    all_x = []
    all_y = []
    for name, subdf in df.groupby(["procedure"]):
        subdf = subdf[subdf.run == run]
        xs = []
        ys = []
        # Only look at the first run
        for _, otherdf in subdf.groupby(["sample"]):
            otherdf = otherdf[otherdf.run == run]
            true_val = evaporated_df[evaporated_df["sample"] == otherdf["sample"].values[0]].tds.values[0]
            xs.extend([true_val]*len(otherdf.tds.values))
            ys.extend(otherdf.tds.values)
        taus[name] = kendalltau(xs, ys)
        plt.scatter(xs, ys, marker="x", label=name)
        all_x.extend(xs)
        all_y.extend(ys)
    plt.xlabel("Total Evaporated Solids (%)")
    plt.ylabel("TDS (%)")
    plt.legend(labels=["spoon stir", "stir plate"])
    plt.annotate(
        "Spoon τ = {:.2f}, p = {:.2f}".format(*taus["spoon stir"]),
        (0.55, 0.01),
        xycoords="axes fraction",
    )
    plt.title(f"Run {run} Rank Ordering")
    plt.annotate(
        "Stir Plate τ = {:.2f}, p = {:.2f}".format(*taus["stir plate"]),
        (0.55, 0.06),
        xycoords="axes fraction",
    )
    plt.annotate(
        "τ = {:.2f}, p = {:.2f}".format(*kendalltau(all_x, all_y)),
        (0.55, 0.11),
        xycoords="axes fraction",
    )
    plt.show()
```

## Hypothesis #2: Conclusion

There is a rank ordering to be found within the data, across all 6 samples. However the sample size is small, particularly with the two methods. There is also a lack of comparison to the conventional rank ordering tool, filtered espresso.
