# Evaluating Impact of Stirring of Espresso on Total Dissolved Solids (TDS)

There have been several studies of the precision of TDS refractometers in the context espresso[[1][1]][[2][2]]. These studies have adopted procedures for stirring the espresso upfront using a spoon/glass rod[[1][1]][[2][2]][[5][5]], specifying a number of stirs. This study looks to evaluate the impact the stirring protocol has on the precision of measuring TDS in espresso as well as the accuracy of the TDS measurements.

## Espresso Methodology

### Initial Preparation

1. Calibrate Acaia Lunar/Pearl using method described [here](https://help.acaia.co/hc/en-us/articles/360035746811-How-to-calibrate-your-Lunar-Pearl-Pearl-Model-S-Pearl-2021-or-Pyxis-scale)

### Pulling Shot

1. Weigh out 18.5g of coffee beans using the Acaia Pearl
1. Spray Beans once with small spray bottle that came with grinder
1. Grind beans into Portafilter, pouring beans in all at once
1. Remove excess grounds to achieve 18.0g within basket using a spoon
1. Tap portafilter on workspace twice, flatten bed using 'distribution' tool, then tamp
1. Place portafilter in machine, place cup on scale, tare.
1. Engage machine and scale timer at same time
1. Disengage machine when weight in cup reaches 36.0g
1. Record final weight, pull time of shot as well as date and time
1. Proceed to [measure TDS](#Measuring-TDS) using randomly selected method 

## Measuring TDS

There are three methods with which TDS will be measured for this experiment, they are as follows:

1. Stir espresso with spoon, sample, cool, measure without filtering
2. Stir espresso with stir plate, sample, cool, measure without filtering
3. Weigh total solids as described [here][4]

The third method will be performed for every shot, which is intended as a more accurate measure of the true TDS. while the sampling order of using a spoon to stir vs a stir plate will be decided up front using a Random Number Generator (RNG) to avoid measurement errors caused by sampling sequentially (getting tired, getting sloppy, etc). While samples are collected on different days, all pairs (stir with spoon vs stir plate) are to be collected on the same day.

### Syringe Reuse

Syringes are made up a significant amount of plastic, and performing this experiment to the standards found in a lab would require a lot of syringes. This wasteful for a measurement that has low impact (IE good vs great coffee compared to if a molecule is a drug or not). With that in mind syringes will be reused in a way that attempts to produce defensible values while avoiding too much waste. 

The syringes that collect the samples from the stirred coffee will be reused for the unfiltered TDS testing as well as collecting the sample for measuring the 'true' dissolved solids via evaporation[[4][4]].

There is an additional syringe used to tare the refractometer with distilled water, which will be reused through out the entire experiment (unless tainted in some obvious way).

### Initial Preparation

1. Calibrate Fuzion Pocket Refractometer with provided 50g weight, instructions provided inside of case.

### Collect Samples

1. Perform stirring method selected for shot [Spoon](#Stirring-With-a-Spoon) or [Stir Plate](#Stirring-With-Stir-Plate)
1. [Extract](#Extracting-Sample-Procedure) samples and place into container attempting to get as close to 10g as possible. Weigh and record total weight of sample, label and set aside.
1. Use 3ml syringe to [extract](#Extracting-Sample-Procedure) two 3ml samples, from below crema, place into metal 'shot glass' to [cool](#Cooling-Procedure).
1. Wait 5 minutes for cooling
1. For three separate 1.0ml samples, perform the following steps
    1. Clean refractometer sensor, tare using distilled water
    1. Draw up 1.0ml of liquid
    1. Dispose of first 4 drops
    1. Fill refractometer sensor area
    1. Collect TDS and Temperature (from refractometer) every 30 seconds until TDS converges three times in a row or 10 times, whichever comes first.
1. Place set aside sample with ~10g into oven at 255F (123.89C)
1. Once sample is dry, measure total weight. TDS computed in notebook


###  Stirring With a Spoon

1. Stir shot with a small spoon ten times clockwise and ten times counterclockwise

### Stirring With Stir Plate

1. Add clean stir rod
1. Turn on stir plate to lowest speed that creates clean vortex without splatter
1. Wait 10 seconds
1. Turn off stir plate, wait for liquid to stop spinning.

## Cooling Procedure

1. Use metal shot glasses to cool samples
1. Place samples on laptop cooling pad, turn on full

```python
# Download CSV, won't work on windozes
with NamedTemporaryFile(suffix=".csv") as temp:
    with open(temp.name, "wb") as ofs:
        resp = requests.get(DATA_DOWNLOAD_URL, stream=True)
        for chunk in resp.iter_content():
            ofs.write(chunk)
    df = pd.read_csv(temp.name)
# Clean up the CSV file.
for char in [" ", "(", ")", "%", "/", "\\", '"']:
    df.columns = df.columns.str.strip().str.lower().str.replace(char, "")
df = df[df["grinder"].astype(bool)][df["grindersetting"].astype(bool)]
# Only look at data with TDS values and ratings
df = df.assign(extraction_yield=(df.tds * df.output) / df.coffeegrams)
df.date = pd.to_datetime(df.date, infer_datetime_format=True, utc=True)
df.roastdate = pd.to_datetime(df.roastdate, infer_datetime_format=True, utc=True)
df = df.assign(days_off_roast=df.date - df.roastdate)

evaporated_df = df[~df["containerweight+driedsolids"].isna()]
evaporated_df = evaporated_df.assign(
    tds=np.abs(
        (
            evaporated_df["containerweight+driedsolids"]
            - evaporated_df["containerweight"]
        )
        / evaporated_df.liquidweight
        * 100
    )
)
evaporated_df = evaporated_df.assign(
    extraction_yield=(evaporated_df.tds * evaporated_df.output)
    / evaporated_df.coffeegrams
)
df.drop(df[df.tds.isna()].index, inplace=True)

# Combine the two DataFrames
ts_val = np.zeros_like(df.tds.values)
extract_yield = np.zeros_like(ts_val)
for i, d in enumerate(df["sample"].values):
    ts_val[i] = evaporated_df[evaporated_df["sample"] == d].tds.values[0]
    extract_yield[i] = evaporated_df[
        evaporated_df["sample"] == d
    ].extraction_yield.values[0]
df = df.assign(
    total_solids=ts_val,
    total_solids_extraction=extract_yield,
)
```

```python
def bootstrap_statistic(x, y, func, samples=1000):
    """Naive Bootstrap of a paired function"""
    sample_vals = []
    x = np.asarray(x)
    y = np.asarray(y)
    for _ in range(samples):
        idxs = np.random.randint(0, len(x), len(x))
        sample_vals.append(func(x[idxs], y[idxs]))
    return sample_vals


def max_absolute_deviation(x):
    """What Reference 2 calls variation"""
    return np.abs(np.max(x) - np.min(x))
```

## Evaluate Normality of TDS Measurements

When using mean, std and variance, it is important to ensure that your data is normal. If it isn't normal, will have to use [non-parametric](https://en.wikipedia.org/wiki/Nonparametric_statistics) tests.

```python
print("Testing distributions for Normality")
alpha = 0.01
for name, subdf in df.groupby(["procedure"]):
    subdf.drop(subdf[subdf.tds.isna()].index, inplace=True)
    # Shapiro Wilk test choosen for significance with smaller sample sizes
    wilk_test = shapiro(subdf.tds.values)
    if wilk_test.pvalue >= alpha:
        print(
            name,
            "possible to have been drawn from a normal distribution, p:",
            np.round(wilk_test.pvalue, 3),
        )
    else:
        print(
            name,
            "unlikely to have been drawn from a normal distribution, p:",
            np.round(wilk_test.pvalue, 3),
        )
```

```python
procedures = defaultdict(list)
# Use first run only
for name, subdf in df[df.run == 1].groupby(["sample", "procedure"]):
    subdf.drop(
        subdf[(subdf.tds.isna()) & (subdf.total_solids.isna())].index, inplace=True
    )
    diffs = subdf.tds.values - subdf.total_solids.values
    procedures[name[1]].extend(diffs)
for key, vals in procedures.items():
    plt.hist(vals, bins=30)
    plt.title(f"Residuals of {key}")
    plt.xlabel("TDS - TS")
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Looking only at run 1, looking at all runs turned out to be less meaningful
# due to the trend that subsequent samples appeared to be evaporating and thus higher in TDS.
print("TDS, Run 1")
print("=" * 10)
first_run_df = df[df.run == 1]
for name, subdf in first_run_df.groupby(["procedure", "sample"]):
    subdf.drop(
        subdf[(subdf.tds.isna()) & (subdf.total_solids.isna())].index, inplace=True
    )
    print(
        "Stir Method",
        name[0],
        "Sample",
        "{:<2}".format(name[1]),
        "Evaporated Total Solids",
        "{:<5}".format(np.round(subdf.total_solids.values[0], 2)),
        "Median",
        "{:5}".format(np.round(subdf.tds.median(), 2)),
        "Max Deviation",
        np.round(max_absolute_deviation(subdf.tds.values), 3),
    )
print()
print("Extraction Yield, Run 1")
print("=" * 23)
for name, subdf in first_run_df.groupby(["procedure", "sample"]):
    subdf.drop(subdf[subdf.tds.isna()].index, inplace=True)
    print(
        "Stir Method",
        name[0],
        "Sample",
        "{:<2}".format(name[1]),
        "Evaporated EY",
        "{:<5}".format(np.round(subdf.total_solids_extraction.values[0], 2)),
        "Median",
        "{:<5}".format(np.round(subdf.extraction_yield.median(), 2)),
        "Max Deviation",
        np.round(max_absolute_deviation(subdf.extraction_yield.values), 3),
    )

procedure_deviations = defaultdict(list)
for name, subdf in df.groupby(["procedure", "sample", "run"]):
    subdf.drop(subdf[subdf.tds.isna()].index, inplace=True)
    procedure_deviations[name[0]].append(max_absolute_deviation(subdf.tds.values))
keys = list(procedure_deviations.keys())

plt.hist([procedure_deviations[key] for key in keys], bins=20, label=keys, alpha=0.25)
plt.legend()
plt.title("Max Deviation of All Runs by Procedure")
plt.xlabel("Max Deviation")
plt.ylabel("Frequency")
plt.show()

both = True
for key in keys:
    wilk_test = shapiro(procedure_deviations[key])
    if wilk_test.pvalue < 0.5:
        both = False
        break
if both:
    print("Both Distributions are normal, performing t-test")
    ttest = ttest_ind(
        procedure_deviations[keys[0]], procedure_deviations[keys[1]], equal_var=False
    )
    if ttest.pvalue > 0.05:
        print(
            "Unlikely that distributions are different, p:", np.round(ttest.pvalue, 3)
        )
    else:
        print(
            "Procedures produce significantly different values, p:",
            np.round(ttest.pvalue, 3),
        )
else:
    print("Distributions are not likely normal, performing KS 2 Sample Test")
    ks_test = ks_2samp(procedure_deviations[keys[0]], procedure_deviations[keys[1]])
    if ks_test.pvalue > 0.05:
        print(
            "Samples are likely to be from the same distribution, p:",
            np.round(ks_test.pvalue, 3),
        )
    else:
        print("Significantly different distribution, p:", np.round(ks_test.pvalue, 3))
```

## Hypothesis #1: Conclusion

## Hypothesis #2: Samples Evaporate Over time

While measuring one sample, the following sample will have a higher TDS due to evaporation. Implies that samples collected and tested sooner might be more 'accurate' for computing Extraction Yield. 

```python
taus = []
rs = []
for name, subdf in df.groupby(["procedure", "sample"]):
    subdf.drop(subdf[subdf.tds.isna()].index, inplace=True)
    plt.scatter(
        subdf.run.values,
        subdf.tds.values,
        label=f"Method {name[0]}, Sample {name[1]}",
        marker="x",
        alpha=0.50,
    )
    taus.append(kendalltau(subdf.run.values, subdf.tds.values))
    rs.append(pearsonr(subdf.run.values, subdf.tds.values)[0] ** 2)
    poly_fit = np.poly1d(np.polyfit(subdf.run.values, subdf.tds.values, 1))
    plt.plot(
        subdf["run"].unique(),
        poly_fit(subdf["run"].unique()),
        linestyle="--",
        alpha=0.25,
    )
plt.title("Impact of Cooling time on TDS")
plt.ylabel("TDS (%)")
plt.xlabel("Run")
plt.xticks(np.arange(1, 4, 1.0))
plt.annotate(
    "Average Kendall Tau = {:.2f}".format(np.mean([x[0] for x in taus])),
    (0.6, -0.22),
    xycoords="axes fraction",
)
plt.annotate(
    "Average Pearson $r^2$ = {:.2f}".format(np.mean(rs)),
    (0.6, -0.27),
    xycoords="axes fraction",
)
plt.show()
```

## Hypothesis #2: Conclusion

There is definitely a trend to indicate that as samples cool for longer periods the TDS increases, likely due to evaporation. This results in a higher TDS measurement. Likely indicates that running multiple samples in this fashion is not useful for comparing variance, without some means to counteract the evaporation.

### Take Away

* Use only the first run of each sample for further analysis. It is also impractical to expect 30 minutes of cooling when testing TDS.

## Hypothesis #3: Unfiltered TDS provides ranks ordering

While unfiltered measurements of espresso do not provide the most precise reading [[1](https://socraticcoffee.com/2015/07/measuring-total-dissolved-solids-a-refractometer-comparison-part-iii-espresso-filters/)], rank ordering would seem to be what matters for dialing in coffee. Despite the lack of precision, does measuring the unfiltered espresso correctly identify a sample x as lower or higher in tds than a sample y?

```python
# Only look at the first run
for run in [1]:
    taus = {}
    maes = {}
    pears = {}
    spears = {}
    all_x = []
    all_y = []
    for name, subdf in df.groupby(["procedure"]):
        subdf = subdf[subdf.run == run]
        xs = []
        ys = []
        for sample, otherdf in subdf.groupby(["sample"]):
            marker = "x"
            if name != "stir plate":
                marker = "^"
            otherdf = otherdf[otherdf.run == run]
            xs.extend(otherdf.total_solids.values)
            ys.extend(otherdf.tds.values)
        taus[name] = kendalltau(xs, ys)
        maes[name] = mean_absolute_error(xs, ys)
        pears[name] = pearsonr(xs, ys)
        spears[name] = spearmanr(xs, ys)
        plt.scatter(xs, ys, marker=marker, label=name)
        all_x.extend(xs)
        all_y.extend(ys)
    all_x = np.asarray(all_x)
    all_y = np.asarray(all_y)
    plt.xlabel("Total Solids (%)")
    plt.ylabel("TDS (%)")
    model_stats = {
        "Spoon œÑ = {:.2f}": [taus["spoon stir"][0]],
        "Stir Plate œÑ = {:.2f}": [taus["stir plate"][0]],
        "œÑ = {:.2f}": [kendalltau(all_x, all_y)[0]],
        "Spoon Pearson $r^2$: {:.2f}": [pears["spoon stir"][0] ** 2],
        "Stir Plate Pearson $r^2$: {:.2f}": [pears["stir plate"][0] ** 2],
        "Pearson $r^2$: {:.2f}": [pearsonr(all_x, all_y)[0] ** 2],
        "Spoon Spearman $r^2$: {:.2f}": [spears["spoon stir"][0] ** 2],
        "Stir Plate Spearman $r^2$: {:.2f}": [spears["stir plate"][0] ** 2],
        "Spearman $r^2$: {:.2f}": [spearmanr(all_x, all_y)[0] ** 2],
        "Spoon mae: {:.2f}": [maes["spoon stir"]],
        "Stir plate mae: {:.2f}": [maes["stir plate"]],
        "mae: {:.2f}": [mean_absolute_error(all_x, all_y)],
    }
    for i, (name, stat) in enumerate(model_stats.items()):
        if i < len(model_stats) // 2:
            x_val = 0.0
        else:
            x_val = 0.55
            i -= len(model_stats) // 2
        plt.annotate(
            name.format(*stat), (x_val, -0.18 - 0.05 * i), xycoords="axes fraction"
        )
    res = theilslopes(all_y, all_x, 0.9)
    lsq_res = linregress(all_x, all_y)
    plt.plot(all_x, res[1] + res[0] * all_x, label="Theil-Sen Fit", alpha=0.5, c="g")
    plt.plot(
        all_x,
        lsq_res[1] + lsq_res[0] * all_x,
        label="Least Squares Fit",
        c="black",
        alpha=0.5,
    )
    plt.grid(True, linestyle="--")
    plt.title(f"TDS's relation to Total Solids")
    plt.legend()
    plt.show()
```

## How reliable are the correlations?

Looking at only 20 samples for correlation on their own is not very robust. Investigate the CI using bootstrapping to determine how reliable the correlations are.

```python
metrics_and_intervals = defaultdict(dict)
quantiles = [0.025, 0.975]
num_samples = 10000
for stat_name, stat in [
    ("Kendall Tau", kendalltau),
    ("Spearman", spearmanr),
    ("Pearson", pearsonr),
]:
    all_xs = []
    all_ys = []
    for name, subdf in df[df.run == 1].groupby(["procedure"]):
        xs = []
        ys = []
        for _, otherdf in subdf.groupby(["sample"]):
            xs.extend(otherdf.total_solids.values)
            ys.extend(otherdf.tds.values)
        vals = bootstrap_statistic(xs, ys, stat, samples=num_samples)
        reference_val, _ = stat(xs, ys)
        if stat_name == "Kendall Tau":
            dist = [val[0] for val in vals]
            label = stat_name
        else:
            dist = [val[0] ** 2 for val in vals]
            label = stat_name + " $r^2$"
            reference_val = reference_val**2
        plt.hist(dist, bins=100, label=label)
        low, high = np.quantile(dist, quantiles)
        plt.axvline(reference_val, label="Reference", c="gray", linestyle="--")
        plt.axvline(low, label="2.5% Quantile", c="green")
        plt.axvline(high, label="97.5% Quantile", c="red")
        plt.ylabel("Frequency")
        plt.xlabel(label)
        plt.title(f"Bootstrapping {label}: {name}")
        plt.legend()
        plt.show()
        all_xs.extend(xs)
        all_ys.extend(ys)
        metrics_and_intervals[f"{label} {name}"].update(
            {
                "low": low,
                "high": high,
                "ref": reference_val,
                "xs": np.asarray(xs),
                "ys": np.asarray(ys),
                "stat": stat,
            }
        )
    reference_val, _ = stat(all_xs, all_ys)
    vals = bootstrap_statistic(all_xs, all_ys, stat, samples=num_samples)
    if stat_name == "Kendall Tau":
        dist = [val[0] for val in vals]
        label = stat_name
    else:
        dist = [val[0] ** 2 for val in vals]
        label = stat_name + " $r^2$"
        reference_val = reference_val**2
    low, high = np.quantile(dist, quantiles)
    metrics_and_intervals[label].update(
        {
            "low": low,
            "high": high,
            "ref": reference_val,
            "xs": np.asarray(all_xs),
            "ys": np.asarray(all_ys),
            "stat": stat,
        }
    )
    plt.axvline(reference_val, label="Reference", c="gray", linestyle="--")
    plt.axvline(low, label="2.5% Quantile", c="green")
    plt.axvline(high, label="97.5% Quantile", c="red")
    plt.hist(dist, bins=100, label=label)
    plt.ylabel("Frequency")
    plt.xlabel(label)
    plt.legend()
    plt.title(f"Bootstrapping {label}, All Samples")
    plt.show()
```

The previous plots display the individual metrics collected by naive bootstrapping for 10000 rounds. It provides us with confidence intervals as well as a visual method to ensure that our reference metric is close to the center of the distribution, an indication that the reference is not random chance. Confidence intervals from bias-corrected and accelerated (BCa) [Bootstrapping](https://en.wikipedia.org/wiki/Bootstrapping_(statistics)#Methods_for_bootstrap_confidence_intervals) are more robust, but its easier to visualize the naive bootstrap.

```python
for key, val in metrics_and_intervals.items():
    print(key)
    print("-" * len(key))
    if "Tau" not in key:
        boot = bootstrap(
            (val["xs"], val["ys"]),
            lambda x, y: val["stat"](x, y)[0] ** 2,
            n_resamples=num_samples,
            paired=True,
            vectorized=False,
        )
    else:
        boot = bootstrap(
            (val["xs"], val["ys"]),
            lambda x, y: val["stat"](x, y)[0],
            n_resamples=num_samples,
            paired=True,
            vectorized=False,
        )
    print(
        "Reference",
        np.round(val["ref"], 3),
    )
    print(
        "Naive Bootstrap 95% CI",
        np.round([val["low"], val["high"]], 3),
    )
    print(
        "Bias Correct and Accelerated Bootstrap 95% CI",
        np.round([boot.confidence_interval.low, boot.confidence_interval.high], 3),
    )
    print()
```

## Hypothesis #3: Conclusion

There is a rank ordering to be found within the data. However the sample size is small, particularly when partitioning by stir method. As it seems that there is good evidence that the different stirring methods made no impact, it seems safe to combine the data together which provides some usable signal. 

Bootstrapping shows that Pearson $r^2$, when all samples combined, stays around 0.7, which would seem to indicate that 70% of the variation in the TS measurement explains the change in the TDS measurement. Confidence for this is relatively low with such a small sample size, but since all three statistics show a correlation it seems to indicate that testing unfiltered shots either stirring with a spoon or a stir plate gives signal when considering rank ordering.

There is a lack of comparison to the conventional rank ordering tool, filtered espresso. Which could have better correlation thanks to a higher precision, though another experiment would need to be run to evaluate the correlation in that scenario.

## Investigation: Which Collection Procedure Produces Most Reliable Results

In this experiment there were 10 samples collected, one every 30 second for each run. It is unclear which sample you should take and record if someone asks for the TDS value of a shot. Is it the first measurement, the third as [Atago](https://www.atago.net/ru/pdf/tanpin-catalog/pal-coffee_a5_en.pdf) suggests, the mean of the first three or is it perhaps the sample that is measured at the temperature closest to the temperature at which the refractometer was tared?

```python
def select_first(x, **kwargs):
    return np.asarray([z[0] for z in x])


def select_mean(x, **kwargs):
    return np.asarray([np.mean(z) for z in x])


def select_median(x, **kwargs):
    return np.asarray([np.median(z) for z in x])


def select_third(x, **kwargs):
    return np.asarray([z[2] for z in x])


def select_mean_of_first_three(x, **kwargs):
    return np.asarray([np.mean(z[:3]) for z in x])


def select_median_of_first_three(x, **kwargs):
    return np.asarray([np.median(z[:3]) for z in x])


def select_smallest_temp_diff(x, tare_temp=None, temps=None):
    if tare_temp is None or temps is None:
        raise ValueError("Must provide temps and tare temp")
    lowest_temps = []
    for subx, tare, temps in zip(x, tare_temp, temps):
        diff = np.abs(temps - tare)
        lowest_temps.append(subx[np.argmin(diff)])
    return np.asarray(lowest_temps)


comps = [
    ("first sample", select_first),
    ("mean of 10 samples", select_mean),
    ("median of 10 samples", select_median),
    ("third sample", select_third),
    ("mean of three", select_mean_of_first_three),
    ("median of three", select_median_of_first_three),
    ("nearest to tare temp", select_smallest_temp_diff),
]

xs = []
ys = []
temps = []
tare_temps = []
for _, otherdf in df[df.run == 1].groupby(["sample"]):
    xs.append(otherdf.total_solids.values[0])
    ys.append(otherdf.tds.values)
    temps.append(otherdf.sampletemp.values)
    tare_temps.append(otherdf.refractometertaretemp.values)

stat_func = lambda x, y: pearsonr(x, y)[0] ** 2


for name, meth in comps:
    print(name)
    print("-" * len(name))
    y = meth(ys, tare_temp=tare_temps, temps=temps)
    ref = stat_func(xs, y)
    boot = bootstrap(
        (xs, y), stat_func, n_resamples=num_samples, paired=True, vectorized=False
    )
    low = boot.confidence_interval.low
    high = boot.confidence_interval.high
    print("Ref", np.round(ref, 3), "BCa 95% CI", np.round([low, high], 3))
    print()
```

## Investigation Summary:

There doesn't appear to be any stand out value when it comes to which samples should be selected as the definitive TDS value to assign to a cup. That is promising though, since it indicates that there isn't something hidden in the data (perhaps you could find something with enough hand waving, but this is about finding useful approaches to selecting data) that would provide more signal. It also validates the previous analysis's conclusion that the correlation does not change dramatically with how you subsample (via bootstrapping) the data.

## Investigation: How Much Does Error Impact Correlation?

How well can TDS measurements perform if we assume certain levels of error within our experimental data? The Atago reports precision that is within 0.15, if we [model this error](http://practicalcheminformatics.blogspot.com/2019/07/how-good-could-should-my-models-be.html) how high of a correlation can we achieve?

Note that the Atago [claims](https://www.atago.net/ru/pdf/tanpin-catalog/pal-coffee_a5_en.pdf) error can be as high as 0.5 for unfiltered espresso.

```python
x = np.linspace(df.tds.values.min(), df.tds.values.max(), 200)
y = np.copy(x)
fig, axs = plt.subplots(3, 2, figsize=(6, 8))
err_folds = [1, 2, 3, 4, 5, 10]
precision = 0.15
axs = axs.reshape(-1)
for i, err in enumerate(err_folds):
    err_add = np.random.normal(0, precision * err, size=y.shape)
    axs[i].scatter(x, y + err_add, s=2)
    axs[i].set_title(f"{err} Fold Error")
    axs[i].set_xlabel("TDS (%)")
    if i % 2 == 0:
        axs[i].set_ylabel("TDS + Error (%)")
fig.suptitle("Modeling Error in Precision")
fig.tight_layout()
plt.show()
```

```python
metric_func = lambda x, y: pearsonr(x, y)[0] ** 2
for refract, precision in [("Atago", 0.15), ("VST V3", 0.01)]:
    cor_list = []
    for fold in err_folds:
        for _ in range(1000):
            error = np.random.normal(0, fold * precision, size=y.shape)
            cor_list.append(
                [metric_func(x, y + error), f"Error: {fold*precision:.2f}\n{fold} Fold"]
            )
    error_df = pd.DataFrame(cor_list, columns=["r2", "fold_error"])
    axes = error_df.boxplot("r2", "fold_error")
    axes.set_xlabel("Fold")
    axes.set_ylabel("Pearson $R^2$")
    axes.set_title("")
    axes.tick_params(labelrotation=50)
    fig = axes.get_figure()
    fig.suptitle(f"{refract} - Base Precision {precision}")
```

## Investigation Summary

As one would expect, the VST III refractormeter behaves much better even if the error is increased, which makes sense as it claims to be 15x more precise. On the other hand, even with an error 3x greater than what is suggested by Atago (for brewed coffee), there is still a correlation of pearson $r^2$ ~= 0.7. That 3x point of 0.45 error is aproximately where unfiltered espresso would lie based on Atago's claims, which suggests that, assuming no other factors, that the Atago should still be capable of providing decent rank ordering without filtering.

## Hypothesis #4: Difference in temperature from taring impacts performance

When a refractometer is tared, it is done so with room temperature distilled water. Coffee samples are cooled in an attempt to bring them to room temperature with the intention of bringing it to the same temperature as the distilled water. It would seem that the closer the sample is to the temperature at which the refractometer was tared for the sample would be the most accurate.

```python
df_copy = df.copy()
taus = []
for (sample, run), subdf in df_copy.groupby(["sample", "run"]):
    delta_tds = subdf.tds.values - subdf.total_solids.values[0]
    delta_temps = subdf.sampletemp.values - subdf.refractometertaretemp.values
    taus.append(kendalltau(delta_temps, delta_tds))
    marker = "x"
    if run == 1:
        marker = "x"
    elif run == 2:
        marker = "o"
    elif run == 3:
        marker = "^"
    plt.scatter(
        delta_temps, delta_tds, label=f"Sample {sample}, Run {run}", marker=marker
    )
plt.ylabel("TDS - TS")
plt.xlabel("Temperature Delta")
plt.title(f"Impact of temperature difference on TDS Measurement")
plt.axvline(0.0, linestyle="--", color="gray")
plt.annotate(
    "Average Kendall Tau = {:.2f}".format(np.mean([x[0] for x in taus])),
    (0.6, -0.22),
    xycoords="axes fraction",
)
plt.grid(True, linestyle="--")
plt.show()
```

## Hypothesis #4: Conclusion


There appears to be a negative trend that as temperature increases from the temperature at which the sensor was tared the TDS value decreases. This shows no sign that the values sampled at the same temperature as the tared value will be any more accurate.
